# Column description and understanding

The scoring is based on ones perception about how one pictures his perfect life and ones current state with respect to the ideal life on a scale of 0-10 with 0 being the worst and 10 being the best.


1. Country - Name of the Country


2. Region - Continent or the Region in which the country is located


3. Happiness Rank - Rank of the Country based on the happiness score


4. Happiness Score - Measure of the happiness the people


5. Economy (GDP per Capita) - It is the measure of GDP per capita. It is a measure of determining the Purchasing Power Parity.


6. Family - It is a measure of Support from Family, Relatives and Friends during difficult situations.


7. Health (Life Expectancy) - Measure of health parameters and longetivity of life.


8. Freedom - It is a measure of Freedom of choices.


9. Trust (Government Corruption) - It is a measure of Trust in the government. It is a measure of degree of corruption in the governement.


10. Generosity - It is a measure of Generosity or Charity by an Individual of the Country.


11. Dystopia Residual - It is the measure of unexplained parameters of the Happiness score.


In this Happiness Score is the Label.

Country and Region are not inputs and are just for recognition.

Happiness rank is the rank based on Happiness Score. Hence it is neither a Feature nor a Label.

Following are the Features of the Data Set:

1. Economy
2. Family
3. Health
4. Freedom
5. Trust
6. Generosity

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics 
```

```python
lr = LinearRegression()
```

```python
x = df.drop(columns=['Happiness Score'])
x
```

```python
y = df['Happiness Score']
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.25, random_state = 355)
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
met = metrics.mean_squared_error(y_pred,y_test)
met
```
